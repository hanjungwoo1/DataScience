```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_df = train_df[train_df["가입자 일련번호"] % 10 == 5]
test_df
```

```python
train_df = train_df[~(train_df["가입자 일련번호"] % 10 == 5)]
train_df
```

```python
train_df.describe()
```

```python
train_df = train_df.drop("기준년도", axis = 1)
train_df = train_df.drop("데이터 공개일자", axis = 1)

train_df.describe()
```

```python
train_df.head()
```

```python
train_df.isnull().sum()
```

```python
look_df = train_df[["체중(5Kg 단위)", "신장(5Cm단위)", "허리둘레", "수축기 혈압", "이완기 혈압", "식전혈당(공복혈당)", "총 콜레스테롤", "혈색소", "혈청크레아티닌"]]
```

```python
look_df = look_df.corr()
```

```python
plt.figure(figsize=(12,12))
plt.rc('font', family='Malgun Gothic')
sns.heatmap(look_df, vmax =1, vmin = -1, center = 0, annot = True, cmap="Blues")
```

```python
waist_mean = train_df[["체중(5Kg 단위)", "허리둘레"]].groupby("체중(5Kg 단위)").mean().reset_index()
```

```python
waist_mean.columns = ["체중(5Kg 단위)", "Waist"]
```

```python
waist_mean
```

```python
train_df = pd.merge(train_df, waist_mean, on="체중(5Kg 단위)", how="left")
train_df
```

```python
train_df.loc[(train_df["허리둘레"].isnull()), "허리둘레"] = train_df["Waist"]

train_df
```

```python
train_df.isnull().sum()
```

```python
look_df = train_df[["성별코드", "연령대 코드(5세단위)", "시력(좌)", "시력(우)", "청력(좌)", "청력(우)"]]
look_df = look_df.corr()
               
plt.figure(figsize=(12,12))
plt.rc('font', family='Malgun Gothic')
sns.heatmap(look_df, vmax =1, vmin = -1, center = 0, annot = True, cmap="Blues")
```
